# 1 - to do list

```python
def main():
    tasks = []

    while True:
        print("\nTo-Do List")
        print("=============================")
        print("1. Add Task")
        print("2. Show Tasks")
        print("3. Mark Task as Done")
        print("4. Exit")

        choice = input("Enter Your Choice: ")

        if choice == '1':
            num_tasks = int(input("How many tasks do you want to add? "))
            for _ in range(num_tasks):
                task_description = input("Enter the task description: ")
                tasks.append({"task": task_description, "done": False})
                print("Task added.")
        
        elif choice == '2':
            if not tasks:
                print("No tasks available.")
            else:
                print("Tasks:")
                for index, task in enumerate(tasks):
                    status = "Done" if task["done"] else "Not Done"
                    print(f"{index + 1}. {task['task']} - {status}")
        
        elif choice == '3':
            index_task = int(input("Enter the task number to mark as done: ")) - 1
            if 0 <= index_task < len(tasks):
                tasks[index_task]["done"] = True
                print("Task marked as done!")
            else:
                print("Invalid task number.")
        
        elif choice == '4':
            print("Exiting To-Do list, goodbye ðŸ‘‹ðŸ‘‹")
            break
        
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

# 2 - calculator

```python
def calc (x,y, operator):
    if operator == 'add':
        return x + y
    elif operator == 'subtract':
        return x - y
    elif operator == 'multiply':
        return x * y
    elif operator == 'divide':
        if y == 0:
            return("opss, cannot divide by zero!")
        return x / y
    else:
        return('Invalid')
    
def main():
    print('Welcome, to Simple Calculator!')
        
        
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))
        
    print("\nWhat operation would yoy like to prefer? ")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiply")
    print("4. Divide")
        
    choice = input("Enter your choice (1,2,3,4): ")
        
    operators = {'1':'add', '2':'subtract', '3':'multiply', '4':'divide'}
        
    operator = operators.get(choice, None)
        
    if operator:
        result = calc(num1, num2, operator)
        print(f"The result is: {result}")
    else:
        print("Sorry, Invalid choice please run the code again")
            
if __name__ == "__main__":
    main()
            
    
```

# 3 - Password generator

```python
import random
import string
#we will use string for lower and upper case alphabets and even digit etc.

#by using func we ask user what the min length is and what they want number,speacialchar etc
def generate_password(min_lenght, number = True, special_characters = True):
    letters = string.ascii_letters
    digits = string.digits
    special = string.punctuation
    #this will print for yoy all the char in lower,upper,number and special char
    
    characters = letters
    if number:
        characters += digits
    if special_characters:
        characters += special
        
    pwd = ""
    meets_criteria = False
    has_number = False
    has_special = False
    
    while not meets_criteria or len(pwd) < min_lenght:
        new_char = random.choice(characters)
        pwd += new_char
        
        if new_char in digits:
            has_number = True
        elif new_char in special:
            has_special = True
            
        meets_criteria = True
        if number:
            meets_criteria = has_number
        if special_characters:
            meets_criteria = meets_criteria and has_special
    return pwd

min_lenght = input("Enter the minimum length: ")
pwd = generate_password(10)
print(pwd)
```

# 4 - Rock-Paper-Scissors

```python

import random

item_list = ["rock", "paper", "scissors"]

def comp_choice():
    return random.choice(item_list)

def show_winner(user_move, comp_move):
    if user_move == comp_move:
        return "It's a tie!"
    elif (user_move == "rock" and comp_move == "scissors") or \
         (user_move == "scissors" and comp_move == "paper") or \
         (user_move == "paper" and comp_move == "rock"):
        return "You Win!"
    else:
        return "Computer Wins!"

def score_board():
    user_score = 0
    comp_score = 0
    rounds_played = 0
    max_rounds = 5
    
    while rounds_played < max_rounds:
        user_input = int(input("Choose 1 for rock, 2 for paper, 3 for scissors (or 'quit' to exit): "))
        if user_input == 'quit':
            print("Goodbye, see you next time!")
            break
        
        if user_input not in [1,2,3]:
            print("Error. Please choose again.")
            continue
            
        user_move = item_list[int(user_input) - 1]
        comp_move = comp_choice()
        print(f"Computer chose: {comp_move}")
        
        result = show_winner(user_move, comp_move)
        print(result)
        
        if result == "You Win!":
            user_score += 1
        elif result == "Computer Wins!":
            comp_score += 1
        
        rounds_played += 1
        print(f"Score - You: {user_score}, Computer: {comp_score}")
        
        if rounds_played < max_rounds:
            play_again = input("Do you want to play again? (yes/no): ").lower()
            if play_again != "yes":
                print("Thanks for playing!")
                break
    
    if rounds_played == max_rounds:
        if user_score > comp_score:
            print("You won the game!")
        elif user_score < comp_score:
            print("Computer won the game!")
        else:
            print("The game is a tie!")
        print(f"Final Score - You: {user_score}, Computer: {comp_score}")

if __name__ == "__main__":
    print("Welcome to Rock-Paper-Scissors!")
    score_board()

```

# 5 - contact book

```python
#learn this new tabulate library from google
from tabulate import tabulate

# dictionary holds key value pair so i have taken empty dictionary
contact = {}

def display_contact():
    if not contact:
        print("Empty contact book")
    else:
        table = []
        for key, value in contact.items():
            table.append([key, value["Phone"], value["Email"], value["Address"]])
        headers = ["Name", "Contact Number", "Email", "Address"]
        print(tabulate(table, headers, tablefmt="grid"))

# menu that display a choice what you want with using while loop
# and inside the while loop i am goin to use (if else statement for six operation)

while True:
    choice = int(input("\n1. Add new contact \n2. Search contact \n3. Display contact \n4. Update contact \n5. Delete contact \n6. Exit, Good bye: \n"))

    if choice == 1:
        name = input("Enter the contact name: ")
        phone = input("Enter the Phone number: ")
        email = input("Enter your email: ")
        address = input("Enter the address: ")
        contact[name] = {"Name": name, "Phone": phone, "Email": email, "Address": address}
        
    elif choice == 2:
        search_name = input("Enter the contact name: ")
        if search_name in contact:
            print(f"{search_name}'s contact number is {contact[search_name]['Phone']}, email is {contact[search_name]['Email']}, address is {contact[search_name]['Address']}")
        else:
            print("Name not found in contact book: ")
            
    elif choice == 3:
        if not contact:
            print("Empty contact book: ")
        else:
            display_contact()
            
    elif choice == 4:
        update_contact = input("Enter the contact to be updated: ")
        if update_contact in contact:
            phone = input("Enter new mobile number: ")
            email = input("Enter new email: ")
            address = input("Enter new address: ")
            contact[update_contact].update({"Phone": phone, "Email": email, "Address": address})
            print("Contact updated")
            display_contact()
        else:
            print("Name is not found in contact book")
            
    elif choice == 5:
        del_contact = input("Enter the contact you want to delete: ")
        if del_contact in contact:
            confirm = input("Do you want to delete this contact yes/no? ")
            if confirm.lower() == 'yes':
                contact.pop(del_contact)
            display_contact()
        else:
            print("Name is not found in contact book")
            
    elif choice == 6:
        print("Good bye!")
        break
        
    else:
        print("Invalid choice! PleaseÂ tryÂ again.")
```
